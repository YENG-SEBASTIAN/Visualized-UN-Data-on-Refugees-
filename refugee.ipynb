```python
import pandas as pd

df = pd.read_csv('https://docs.google.com/spreadsheets/d/1omEKUsryhkN28bQdVnwWKbZxQUaWZDyqGuK0YVGcKQI/export?format=csv')

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visualization style to 'whitegrid' for better readability
sns.set(style='whitegrid')

# Create a figure with specified size and white background
plt.figure(figsize=(10, 6), facecolor='white')

# Identify the top 10 countries based on total administrative cost
# Assuming 'df' is your DataFrame and it contains a column for costs
top_countries = df.nlargest(10, 'Total Administrative Cost for Host Country')

# Plot the data as a bar chart with 'skyblue' color
top_countries.plot(kind='bar', color='skyblue')

# Set the title and labels for the plot
plt.title('Top 10 Countries by Total Administrative Cost for Host Country (2021)')
plt.xlabel('Country or Territory')
plt.ylabel('Total Administrative Cost (in billion USD)')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Adjust layout for a tight fit
plt.tight_layout()

# Save the plot to a PNG file
plt.savefig('top_countries_admin_cost.png')

# Display the plot
plt.show()

# Confirm the visualization has been generated and saved
print('Visualization generated and saved as top_countries_admin_cost.png.')
```
